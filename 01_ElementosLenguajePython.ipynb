## Intrucciones de Control

### Condicionales

```python
# Dado un nro aleatorio, indicar si es par o no
import random
n=random.randint(1,100)
resultado="IMPAR"
if n%2==0:
  resultado="PAR"
print(f"El numero {n} es {resultado}")
```

```python
# Al lanzar un dado indicar si obtuvo el maximo puntaje o sigue jugando
d=random.randint(1,6)
if d==6:
  mensaje="Máximo Puntaje ...!!!!"
else:
  mensaje="Sigue intentando"
print(f"resultado: {d} ({mensaje})")
```

```python
# Condicional en una sola linea
d=random.randint(1,6)
f"Máximo Puntaje({d}) ...!!!!" if d==6 else f"Sigue intentando({d})"
```

### Repetitivas

```python
# Cuando sabemos el nro de repeticiones "for"
for i in range(1,6): # repite desde 1 hasta 5
  print(i)
```

```python
"""
Para un número determinado de alumnos, 
indicar el promedio de sus notas (suma de notas/cantidad de alumnos)
(se debe solicitar las notas de cada alumno)
"""
n=int(input("Ingrese el nro de alumnos: "))
acumulador=0
for i in range(1,n+1):
  nota=float(input(f"Ingrese la nota[{i}]: "))
  acumulador=acumulador+nota
prom=acumulador/n
prom=round(prom,2)
print(f"El promedio de las {n} notas es {prom}")
```

```python
# Cuando NO sabemos el nro de repeticiones "while"
x=0
while x<5:
  print(x)
  x=x+1
```

```python
# Lanzar un dado, varias veces hasta obtener el maximo puntaje, mostrar cada lanzamiento
contador=0
d=0
while d!=6: # Mientras d sea diferente de 6 (!= Operador de diferente)
  d=random.randint(1,6)
  contador=contador+1
  print(f"Lanzamiento[{contador}]: {d}")
```

**Aplicación**: Para la siguiente lista [15, 14, 12, 5, 18, 20, 16] imprimir los elementos de indice par exceptuando el ultimo elemento

```python
i=0
while notas[i]!=notas[-1]: # Mientras el elemento actual sea diferente del ultimo elemento hacer:
  if i>0:
    if i%2==0: # Si es PAR
      print(f"Indice: {i}, Valor: {notas[i]}")
  i=i+1  # Hace que el indice aumente repetición a repetición (iteración a iteración)
```

```python
# ***********Slicing*************
print(notas)
print(notas[3:])    # Del indice 3 en adelante
print(notas[:3])    # Hasta el indice 2
print(notas[2:6])   # Del indice 2 al 5
print(notas[:-1])   # Todos menos el ultimo
```

### Compresion de Colecciones

```python
# LISTAS: [expresion for variable in lista if condicion]
lista=[x**2 for x in range(1,10)]
lista
```

```python
# indicar los elementos de "indice par" de [15, 14, 12, 5, 18, 20, 16]
lista1=[15, 14, 12, 5, 18, 20, 16]
lista2=[v for i,v in enumerate(lista1) if (i%2==0 and i>0)]
lista2
```

```python
# DICCIONARIOS: {expresion-clave:expresion-valor for variables in lista if condicion}
dict_1={palabra:len(palabra) for palabra in ['I', 'love', 'Python']}
dict_1
```
